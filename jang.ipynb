```python
for year in range(2012, 2027):
    
    merged[f"Population_{year}"] = (
        merged["Population"] *
        ((1 + merged["Annual_Growth_Rate"]) ** (year - 2011))
    ).round().astype(int)
```

```python
print(
    merged[[
        "District",
        "Population",
        "Population_2012",
        "Population_2015",
        "Population_2020",
        "Population_2026"
    ]].head()
)
```

```python
def get_population(state, district, year):
    
    state = state.upper()
    district = district.upper()

    row = merged[
        (merged["State"] == state) &
        (merged["District"] == district)
    ]

    if row.empty:
        return "District not found"

    if year == 2011:
        return int(row["Population"].iloc[0])

    return int(row[f"Population_{year}"].iloc[0])
```

```python
get_population(
    "UTTAR PRADESH",
    "DEORIA",
    2025
)
```

```python
def district_forecast(state, district):

    state = state.upper()
    district = district.upper()

    row = merged[
        (merged["State"] == state) &
        (merged["District"] == district)
    ]

    if row.empty:
        return "District not found"

    result = {"2011": int(row["Population"].iloc[0])}

    for year in range(2012, 2027):
        result[str(year)] = int(
            row[f"Population_{year}"].iloc[0]
        )

    return result
```

```python
district_forecast(
    "UTTAR PRADESH",
    "DEORIA"
)
```

```python
import matplotlib.pyplot as plt

years = [2001, 2011]
population = [
    merged["Persons"].sum(),
    merged["Population"].sum()
]

for year in range(2012, 2027):
    years.append(year)
    population.append(
        merged[f"Population_{year}"].sum()
    )

plt.figure(figsize=(14,6))

plt.plot(
    years,
    population,
    marker="o",
    linewidth=2
)

plt.axvline(
    x=2011,
    linestyle="--",
    alpha=0.7,
    label="Forecast Starts"
)

plt.title("India Population Forecast (2001–2026)")
plt.xlabel("Year")
plt.ylabel("Population")
plt.legend()
plt.grid(True)

plt.show()
```
